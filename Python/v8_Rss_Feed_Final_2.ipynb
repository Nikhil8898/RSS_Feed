```python
import pymongo as pym
import pandas as pd
import requests
import nltk
from bs4 import BeautifulSoup
#from bs4 import BeautifulSoup as bs
import urllib.request, urllib.parse, urllib.error
from newspaper import Article 
import schedule
import time
```

```python
#Creating URL Table by inserting the URL when needed
while True:
    ur = {}
    client = pym.MongoClient('mongodb://localhost:27017/')
    db = client["Rss_Feed"]
    rss_feeds_link = db["Rss_Feed_Url"]

    print("Enter choice : \n 1. Insert \n")
    n = int(input())
    if n == 1:
        url = input()
        ur["link"] = url
        rss_feeds_link.insert_one(ur)  
    else:
        break
    
```

```python
def rss_feed_schedule():
    #url=["https://www.coindesk.com/arc/outboundfeeds/rss/?outputType=xml","https://blockchain.news/RSS/","https://news.bitcoin.com/feed/"]
    #Making Connection with mongodb
    try:
        client = pym.MongoClient('mongodb://localhost:27017/')
        #creating db
        db = client["Rss_Feed"]
        #creating tables
        rss_feeds_info = db["Rss_Feed_Info"]
        rss_feeds_link = db["Rss_Feed_Url"]
        url = rss_feeds_link.find()
    except:
        print("Database not Connected")
    for u in url:
        print(u)
        client = pym.MongoClient('mongodb://localhost:27017/')
        
        resp=requests.get(u['link'])
        soup=BeautifulSoup(resp.content,features='xml')
        items=soup.findAll('item')
       
        for item in items:        
            #print(item)
            try:
                a = item.link.text
                article = Article(a)
                article.download()
                article.parse() 
                article.nlp()
            except:
                print("Cannot Parse the Feed")
            qry = {"Title":item.title.text, "Link":item.link.text, "Description":item.description.text,
                   "PubDate":item.pubDate.text, "Keyword":article.keywords}
            qry1 = {"Title":item.title.text, "Link":item.link.text, "PubDate":item.pubDate.text}
            #rss_feeds_info.insert_one(qry)
    
            rss_feeds_info.update_one(qry1,{'$set':qry},upsert=True)
        time.sleep(1)

#Scheduling every 1 hr 
schedule.every(60).seconds.do(rss_feed_schedule)
#schedule.every(1).hour.do(rss_feed_schedule)
while True:
    schedule.run_pending()
    #print("Data")
    time.sleep(1)
    #print("Added")
```
